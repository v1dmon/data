```python
import os
import sys
import pandas as pd

sys.path.append(os.path.abspath(os.path.join('../../analysis')))
```

```python
from analysis.proc.experiment import Experiment

x = Experiment(
    name="Asymmetric Run 1h",
    i1='sockshop-2023-11-05T15-27-22/asym1h/1',
    # i2='sockshop-2023-11-05T15-27-22/asym1h/2',
    # i3='sockshop-2023-11-05T15-27-22/asym1h/3',
)
```

```python
from analysis.plot.figure import Grid
import plotly.graph_objects as go

m1 = x.i1.metric

g = Grid(
    rows=3, cols=2,
    subplot_titles=["Cpu Usage", "Memory Usage", "Time Delta", "Throughput", "Latency"],
    horizontal_spacing=0.05, vertical_spacing=0.1,
)
g.add(go.Scatter(x=m1.cpuUsage.Timestamp, y=m1.cpuUsage.CPUPercUsage, mode="lines"))
g.add(go.Scatter(x=m1.memoryUsage.Timestamp, y=m1.memoryUsage.MemoryPercUsage, mode="lines"))
g.add(go.Scatter(x=m1.timedelta.Timestamp[:11000], y=m1.timedelta.TimeDelta[:11000], mode="lines"))
g.add(go.Scatter(x=m1.throughput.Timestamp, y=m1.throughput.Throughput, mode="lines"))
g.add(go.Scatter(x=m1.latency.Timestamp, y=m1.latency.Latency, mode="lines"))
g.layout(showlegend=False, title='Iteration 1 Metrics', height=800)
g.show()
```
